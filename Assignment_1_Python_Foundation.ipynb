**3. Program untuk Menghitung BMI**




```python
def hitung_bmi(berat_badan, tinggi_badan):
    bmi = berat_badan / ((tinggi_badan /100) ** 2)
    return bmi

try:
    berat_badan = float(input("Masukkan berat badan Anda (kg): "))
    tinggi_badan_str = input("Masukkan tinggi badan Anda (cm): ").replace(',', '.')

    tinggi_badan = float(tinggi_badan_str)

    if berat_badan <= 0 or tinggi_badan <= 0:
        print("Masukkan berat badan dan tinggi badan yang valid (harus lebih dari 0).")
    else:
        bmi = hitung_bmi(berat_badan, tinggi_badan)
        print(f"BMI Anda adalah: {bmi:.6f}")
except ValueError:
    print("Masukkan berat badan dan tinggi badan yang valid (angka).")
```

**5. Status data Frame df_b dengan keterangan:**

Underweight : < 18,5

Normal : 18,5 - 24,9

Overweight : 25,0 - 29,9

Obesity : 30,0 - 34,9

```python
def get_status_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25.0:
        return 'Normal'
    elif 25.0 <= bmi < 30.0:
        return 'Overweight'
    else:
        return 'Obesity'

df1_b['Status BMI'] = df1_b['BMI'].apply(get_status_bmi)
df1_b['Status BMI']
```

```python
df2_b = pd.DataFrame(df1_b['Status BMI'], columns=['Status BMI'])
df2_b
```

**9. Mencari Nilai dari :**
*   Kelompokkan tinggi badan dan berat badan serta cari nilai terendah dan tertingginya.
*   Berapa nilai rata-rata tinggi badan siswa?
*   Berapa rata-rata berat badan siswa?
*   Ada berapa orang jumlah siswa  yang tercatat pada dataframe df?
*   5 orang dengan nilai BMI tertinggi.
*   Ada berapa orang yang tergolong status Underweight, Normal, Overweight, dan Obesity?

```python
#Mengelompokkan tinggi badan dan berat badan Kompi B

grouped_data = df3_b.groupby(['Nama', 'Tinggi Badan (cm)', 'Berat Badan (Kg)']).size().reset_index(name='Jumlah Siswa')
grouped_data
```

```python
#Mencari nilai terendah dan tertinggi dari berat badan dan tinggi badan Kompi B

tinggi_berat_stats = df3_b.groupby(['Tinggi Badan (cm)', 'Berat Badan (Kg)']).size().reset_index(name='Jumlah Siswa')
tinggi_min = tinggi_berat_stats['Tinggi Badan (cm)'].min()
tinggi_max = tinggi_berat_stats['Tinggi Badan (cm)'].max()
berat_min = tinggi_berat_stats['Berat Badan (Kg)'].min()
berat_max = tinggi_berat_stats['Berat Badan (Kg)'].max()

print("Tinggi Badan Terendah:", tinggi_min)
print("Tinggi Badan Tertinggi:", tinggi_max)
print("Berat Badan Terendah:", berat_min)
print("Berat Badan Tertinggi:", berat_max)
```

```python
# Mencari nilai rata-rata dari tinggi badan Kompi B

rata_rata_tb = df3_b['Tinggi Badan (cm)'].mean()
print("Rata-rata Tinggi Badan Siswa:", rata_rata_tb)
```

```python
#Mencari nilai rata-rata dari berat badan Kompi B

rata_rata_bb = df3_b['Berat Badan (Kg)'].mean()
print("Rata-rata Berat Badan Siswa:", rata_rata_bb)
```
